## Data processing for the food2fork project

Using the food2fork API, we were able to retrieve information about ~30000 online recipes. We stored all of this information in a MongoDB database. Before we can do anything with it, we need to parse and clean up the data.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pymongo
import pandas as pd
```

```python
client = pymongo.MongoClient('mongodb://localhost:27017')
dbfood = client.food2forkDB.food2fork
```

Let's create a dataframe with all of the data from the database

```python
df = pd.DataFrame.from_records(dbfood.find())
df.head()
```

```python
df.count()
```

### Social media ranking

```python
df['media_likes'] = df['media'].map(lambda x: x.get('likes', 0))
df['media_tweets'] = df['media'].map(lambda x: x.get('tweets', 0))
df['media_plusses'] = df['media'].map(lambda x: x.get('plusses', 0))
df['media_pins'] = df['media'].map(lambda x: x.get('pins', 0))
df['media_views'] = df['media'].map(lambda x: x.get('views', 0))
```

```python
df['nmedia_likes'] = df['media_likes']/df['media_likes'].sum()
df['nmedia_tweets'] = df['media_tweets']/df['media_tweets'].sum()
df['nmedia_plusses'] = df['media_plusses']/df['media_plusses'].sum()
df['nmedia_pins'] = df['media_pins']/df['media_pins'].sum()
df['nmedia_views'] = df['media_views']/df['media_views'].sum()

df['media_rank'] = (df['nmedia_likes'] + df['nmedia_tweets'] +
                    df['nmedia_plusses'] + df['nmedia_pins'] +
                    df['nmedia_views'])
```

```python
sdf = df.sort_values(by='media_rank', ascending=False)
sdf.head()
```

#### Most popular online recipes

```python
sdf = sdf.reset_index(drop=True)
sdf.loc[:20, ['title', 'media_rank']].set_index('title', drop=True).plot(kind='bar')
```

### Time parsing

Here we will transform the time strings into a number of minutes

```python
from dateutil.parser import parse
import re

def parse_time(string):
    if not isinstance(string, str):
        return
    try:
        time = parse(string)
        return time.hour*60+time.minute
    except ValueError:
        pass
    match = re.match('^PT(([\d\.]+)H)?((\d+)M)?', string)
    if match:
        hrs = float(match.group(2) or 0)
        mins = float(match.group(4) or 0)
        return hrs*60+mins

    pattern = (r'^\s*((?P<hours>\d+)\s*(h|h(ou)?rs?))?[\s,]*'
               r'((?P<minutes>\d+)\s*(m|mins?|minutes))?\s*(plus\s.*)?$')
    match = re.match(pattern, string, re.IGNORECASE)
    if match:
        hrs = int(match.group('hours') or 0)
        mins = int(match.group('minutes') or 0)
        return hrs*60+mins
    
    if '-' in string:
        spl = string.split('-')
        return sum(parse_time(item) for item in spl)
    else:
        print(string)


df['cooktime_mins'] = df['cooktime'].map(parse_time)
df['preptime_mins'] = df['preptime'].map(parse_time)
df['totaltime_mins'] = df['cooktime_mins'].fillna(0) + df['preptime_mins'].fillna(0)
```

```python
df[['cooktime', 'cooktime_mins', 'preptime', 'preptime_mins', 'totaltime_mins']].head(50)
```

```python
diff_df = df[['difficulty', 'totaltime_mins']].dropna()
diff_df = diff_df[(diff_df['totaltime_mins'] > 0.) & (diff_df['totaltime_mins'] < 1000.)]
diff_df.head()
```

#### Total cooking time vs difficulty

```python
fig, ax = plt.subplots()
newdf = None
for j, diff in enumerate(difficulties):
    vals = diff_df.loc[diff_df['difficulty'] == diff, 'totaltime_mins'].values
    newdf1 = pd.DataFrame({diff: vals})
    if newdf is None:
        newdf = newdf1
    else:
        newdf = pd.concat([newdf, newdf1], axis=1)

newdf.boxplot(rot=45, showfliers=False)
plt.ylabel('Total cooking time (in mins)')
```

### Ingredient parsing

We're using machine learning to parse the raw recipe ingredients into actual ingredients and get rid of quantities, units and comments

```python
from food2fork.ingredient_parser import get_ingredient_from_phrase
```

```python
df['ingredients'] = df['raw_ingredients'].map(lambda x: [get_ingredient_from_phrase(item) for item in x])
```

### Find most common ingredients

```python
from collections import Counter
from itertools import combinations

def count_ingredients(values, nsub=2):
    '''returns a counter of paired ingredients
    '''
    counter  = Counter()
    for sub in values:
        if len(values) < 2:
            continue
        sub.sort()
        if '' in sub:
            sub.remove('')
        if nsub == 1:
            iterator = sub
        else:
            iterator = combinations(sub, nsub)
        for comb in iterator:
            counter[comb] += 1
    return counter

counter = count_ingredients(df['ingredients'].values, nsub=1)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lmtzr = WordNetLemmatizer()

counter = count_ingredients(df['ingredients'].values, nsub=1)

def filter_ingredients(ingredients, firstRound=True, counter=counter):
    out = []
    for item in ingredients:
#         print(item, item in counter, counter[item] if item in counter else 0, item in ignore)
        if firstRound:
            condition = counter[item] > 10 and item not in ignore;
        else:
            condition = (counter[item] > 20 and
                         item not in ignore and
                         item in categories and
                         categories[item] != 'other')
        if item in counter and condition:
            lem = lmtzr.lemmatize(item)
            if lem not in ignore:
                if lem in mapping:
                    if mapping[lem] not in ignore:
                        out.append(mapping[lem])
                else:
                    out.append(lem)
#     print(ingredients, out)
    return out

df['curated_ingredients'] = df['ingredients'].map(filter_ingredients)
counter = count_ingredients(df['curated_ingredients'].values, nsub=1)
df['curated_ingredients'] = df['curated_ingredients'].map(lambda x: filter_ingredients(x, firstRound=False, counter=counter))
```

### Most common food combinations

```python
def most_common(values, nsub=2, nitems=50, distribution=None):
    pairs = count_ingredients(values, nsub=nsub)
    unique_ingredients = set([ing for inglist in values for ing in inglist])

    out = {}
    for ing in unique_ingredients:
        out[ing] = []
        count_dist = {}
        if distribution:
            for cat in distribution:
                count_dist[cat] = 0
        cnt = 0
        for (ing1, ing2), number in pairs.most_common():
            if number < 5:
                continue
            ing2add = None
            if ing1 == ing and ing2 != ing:
                ing2add = ing2
            elif ing2 == ing and ing1 != ing:
                ing2add = ing1
            if ing2add:
                if ing == 'okra':
                    print(ing2add, number)
                cat = categories[ing2add]
                if (distribution is not None and
                        cat in distribution and
                        count_dist[cat] < distribution[cat]):
                    out[ing].append(ing2add)
                    count_dist[cat] += 1
                    cnt += 1
                elif distribution is None:
                    out[ing].append(ing2add)
                    cnt += 1
            if cnt == nitems:
                break

    return out

pairings = most_common(df['curated_ingredients'].values)
```

### Recipe types

We will use machine learning to categorize recipes between dessert, drink, ...

```python
subdf = df[['curated_ingredients']]
```

```python
subdf = pd.get_dummies(subdf['curated_ingredients'].apply(pd.Series).stack()).sum(level=0)
```

```python
subdf['title'] = df['title']
subdf['recipe_type'] = ''
subdf.head()
```

```python
sample = subdf.sample(n=500)

j = 0
for index, row in sample.iterrows():
    ans = input(f"{j}, {row['title']}: ")
    if ans == 'd':
        val = 'dessert'
    elif ans == 'k':
        val = 'drink'
    else:
        val = 'main'
    sample.set_value(index,'recipe_type', val)
    j += 1
```

```python
sample
```

```python
len(sample)
```

```python
from sklearn.model_selection import train_test_split

X = sample.iloc[:, :-2]
y = sample.iloc[:, -1]

x_train, x_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
print(f'Training score: {logreg.score(x_train, y_train)}')
print(f'Testing score: {logreg.score(x_test, y_test)}')
```

```python
subdf['recipe_type'] = logreg.predict(subdf.iloc[:, :-2])
```

```python
# Saving the model
import pickle

with open('recipe_type.model', 'wb') as outfil:
    pickle.dump(logreg, outfil)
```

```python
# Loading the model for testing
import pickle

with open('recipe_type.model', 'rb') as fil:
    model = pickle.load(fil)

model.predict(subdf.iloc[:10, :-2])
```
